```python
def show_plan(rdd):
    for x in rdd.toDebugString().decode().split('\n'):
        print(x)
```

```python
rdd = sc.textFile("data/romeo_and_juliet.txt")
```

```python
type(rdd)
```

```python
rdd.take(20)
```

```python
words_fm = rdd.flatMap(lambda line: line.split(" "))
words_fm.take(20)
```

```python
words_m = rdd.map(lambda line: line.split(" "))
words_m.take(20)
```

```python
words = rdd.flatMap(lambda line: line.lower().split(" "))
not_empty = words.filter(lambda x: x != '')
key_values = not_empty.map(lambda word: (word, 1))
counts = key_values.reduceByKey(lambda a, b: a + b)
```

```python
show_plan(counts)
```

# Optimized WordCount

```python
import re
```

```python
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
```

```python
sc = SparkContext(master="local[*]", appName="SparkWC")
```

```python
data = sc.textFile("data/romeo_and_juliet.txt")
data.take(10)
```

```python
words = data.flatMap(normalizeWords)
words.take(10)
```

```python
wordCounts = words.countByValue()
type(wordCounts)
```

```python
for word, count in wordCounts.items():
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print((cleanWord.decode(), count))
```
